```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python

df['age'].fillna(df['age'].mean(), inplace=True)
df.head()
```

```python
df = df[['survived','pclass','age','alive']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
oe = LabelEncoder()
df['live_user'] = oe.fit_transform(df[['alive']]).astype(int)
```

```python
df.head()
```

```python
df = df.drop(columns=['alive'] ,axis=1)
```

```python
df.head()
```

```python
X = df.iloc[:, :3]
y = df.iloc[:, -1:]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2 )
```

```python
import tensorflow
from tensorflow import keras
from keras import  Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(3, activation='relu', input_dim = 3))
model.add(Dense(2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model
```

```python
model.fit(X_train,y_train, epochs=10, batch_size=X.shape[0],validation_split=0.2)
```
